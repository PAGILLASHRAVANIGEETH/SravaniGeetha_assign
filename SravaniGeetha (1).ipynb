```python
new_column_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
iris_data.columns = new_column_names
```

```python
iris_data
```

```python


statistics = {}

for column in iris_data.columns[:-1]:
    #print(column)
    column_stats = {
        'Mean': iris_data[column].mean(),
        'Median': iris_data[column].median(),
        'Mode': iris_data[column].mode()[0],
        'Min': iris_data[column].min(),
        'Max': iris_data[column].max(),
        'Standard Deviation': iris_data[column].std(),
        'Variance': iris_data[column].var()
    }



    statistics[column] = column_stats
    #print(statistics[column],"getttttttttttttttt" )
    #print(statistics)

statistics_df = pd.DataFrame(statistics)
print(statistics_df)
```

# Histograms


```python

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.histplot(iris_data['SepalLength'], kde=True, bins=20)
plt.title('Histogram of Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(iris_data['SepalWidth'], kde=True, bins=20)
plt.title('Histogram of Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(iris_data['PetalLength'], kde=True, bins=20)
plt.title('Histogram of Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(iris_data['PetalWidth'], kde=True, bins=20)
plt.title('Histogram of Petal Width')

plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLength', data=iris_data)
plt.title('Boxplot of Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidth', data=iris_data)
plt.title('Boxplot of Sepal Width by Species')

plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLength', data=iris_data)
plt.title('Boxplot of Petal Length by Species')

plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidth', data=iris_data)
plt.title('Boxplot of Petal Width by Species')

plt.tight_layout()
plt.show()
```

# Pairplot

```python
sns.pairplot(iris_data, hue='Species', markers=["o", "s", "D"])
plt.show()
```

# Heatmap

```python
iris_data.columns
```

```python
numeric_columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
correlation_matrix = iris_data[numeric_columns].corr()



plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Scatter_plot

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLength', y='PetalLength', hue='Species', data=iris_data, palette='deep', style='Species', s=100)
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
plt.show()
```
